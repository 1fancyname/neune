title: The Perceptron as a binary Classifier
---

In the last chapter we learned that we can use the following function to describe a binary classifier:  

$$
y = 
\begin{cases}
1 & \text{if } \mathbf{w} \cdot \mathbf{x} + b \ge 0 \\
0 & \text{otherwise}
\end{cases}
$$
$$
\text{where: } y = \text{target class} \\
\mathbf{w} = \text{weight vector} \\
\mathbf{x} = \text{input vector} \\
b = \text{bias}
$$

If we look at the decision boundary as a simple linear function we can write it like this:

$$
y = m x + c
$$

If we compare this with our equation from the perceptron we can derive that:  
$$y = x_2 \\  
m = -\frac{w_1}{w_2} \\  
c = b$$
